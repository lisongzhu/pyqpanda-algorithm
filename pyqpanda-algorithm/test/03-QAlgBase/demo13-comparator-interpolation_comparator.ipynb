```python
from pyqpanda_alg.QFinance import comparator, grover, QAE, QUBO, QSVD, class_qsvr, QmRMR, class_basic_sparecode
import numpy as np
from pyqpanda3.core import *
import matplotlib.pyplot as plt
```

```python
value = 3.3
m = CPUQVM()
prog = QProg()
prog << X(0) << X(1) << I(2)
# 测试该量子态和3.3的大小比较,是基于插值比较的,3.3看成3和4的叠加,量子态3有一部分大于等于3,另一部分小于3
# 这个方法很奇怪，不好说，没什么用的一个接口，相当于可以小小数比较
cir = comparator.interpolation_comparator(value, [0, 1, 2], [3, 4, 5], function='g', reuse=True)
prog << cir
m.run(prog, 1000)
prob_dict_result = m.result().get_prob_dict([5])
print(prob_dict_result)
```
