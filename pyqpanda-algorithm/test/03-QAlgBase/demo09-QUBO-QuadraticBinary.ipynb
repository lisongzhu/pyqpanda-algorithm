```python
import sympy as sp
import numpy as np
from pyqpanda_alg.QFinance import QUBO
from pyqpanda3.core import CPUQVM, QCircuit, QProg, TOFFOLI, X, Z, H, U1, RY
```

```python
x0, x1, x2 = sp.symbols('x0 x1 x2')
function = -0.5 * x0 * x1 - 0.7 * x0 * x1 + 0.9 * x1 * x2 + 1.3 * x0 - x1 - 0.5 * x2
test0 = QUBO.QuadraticBinary(function)
print(test0.function_value([0, 1, 0]))
res0 = test0.qubobytraversal()
print('result of traversal: ', res0)
n_key, n_res = test0.query_qnumber()
print(n_key, n_res)
m = CPUQVM()
q_key = QProg(n_key + n_res).qubits()
print(test0.cir(q_key[:n_key], q_key[n_key:]))
print(test0.cir(q_key[:n_key], q_key[n_key:]).originir())
```
