```python
from pyqpanda_alg.QFinance import comparator, grover, QAE, QUBO, QSVD, class_qsvr, QmRMR, class_basic_sparecode
import numpy as np
from pyqpanda3.core import *
import matplotlib.pyplot as plt
```

```python
'''
需要光滑的振幅加载才有效,例如正态分布,同时可以考虑mode='walsh',和mode='fourier'两种方案
画出来的图差不多就行
Returns:

'''
mu = 0
sigma = 1
x = np.linspace(-3, 3, 2 ** 10)
pdf_normal = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-(x - mu) ** 2 / (2 * sigma ** 2))
# pdf_normal = (1 / (x * sigma * np.sqrt(2 * np.pi))) * np.exp(-(np.log(x) - mu) ** 2 / (2 * sigma ** 2))
ini = pdf_normal / np.linalg.norm(pdf_normal)
# res = class_basic_sparecode.QSpare_Code(ini ** 2, mode='walsh', cut_length=80).Quantum_Res()
res = class_basic_sparecode.QSpare_Code(ini ** 2, mode='walsh', cut_length=20).Quantum_Res()

plt.plot(x, res)
plt.plot(x, ini ** 2)
plt.show()
```

```python
'''
需要光滑的振幅加载才有效,例如正态分布,同时可以考虑mode='walsh',和mode='fourier'两种方案
画出来的图差不多就行
Returns:

'''
mu = 0
sigma = 1
x = np.linspace(-3, 3, 2 ** 10)
pdf_normal = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-(x - mu) ** 2 / (2 * sigma ** 2))
# pdf_normal = (1 / (x * sigma * np.sqrt(2 * np.pi))) * np.exp(-(np.log(x) - mu) ** 2 / (2 * sigma ** 2))
ini = pdf_normal / np.linalg.norm(pdf_normal)
# res = class_basic_sparecode.QSpare_Code(ini ** 2, mode='walsh', cut_length=80).Quantum_Res()
res = class_basic_sparecode.QSpare_Code(ini ** 2, mode='fourier', cut_length=20).Quantum_Res()

plt.plot(x, res)
plt.plot(x, ini ** 2)
plt.show()
```
