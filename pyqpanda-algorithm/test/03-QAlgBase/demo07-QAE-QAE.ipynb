```python
from pyqpanda_alg.QFinance import comparator, grover, QAE, QUBO, QSVD, class_qsvr, QmRMR, class_basic_sparecode
import numpy as np
from pyqpanda3.core import *
import matplotlib.pyplot as plt
```

```python

def create_cir(qlist):
    cir = QCircuit()
    cir << RY(qlist[0], np.pi / 3) << X(qlist[1]).control(qlist[0])
    return cir
W = QAE.QAE(operator_in=create_cir, qnumber=2, epsilon=0.01, res_index=[0, 1], target_state='11').run()
# 这个是测最后两个比特‘11’的振幅
print(W)
# 以上结果都符合预期,没什么好说的,可以改一改线路放进去测一下
```
