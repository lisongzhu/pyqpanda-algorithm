```python
from pyqpanda_alg.QFinance import grover
from pyqpanda3.core import CPUQVM, QCircuit, QProg, TOFFOLI, Z
```

```python
m = CPUQVM()
q_state = QProg(3).qubits()
```

```python
def mark(qubits):
    cir = QCircuit()
    cir << TOFFOLI(qubits[0], qubits[1], qubits[2])
    cir << Z(qubits[2])
    cir << TOFFOLI(qubits[0], qubits[1], qubits[2])
    return cir
prog = grover.amp_operator(q_input=q_state[:2], q_flip=q_state, q_zero=q_state[:2], flip_operator=mark)
print(prog)
print(prog.originir())
```
