```python
from pyqpanda_alg.QAOA import dstate
from pyqpanda3.core import CPUQVM, QProg, draw_qprog
from pyqpanda3.intermediate_compiler import convert_qprog_to_originir
```

```python
n = 4
k = 2

prog = QProg(n)
qubits = prog.qubits()
prog << dstate.prepare_dicke_state(qubits, k)
print(draw_qprog(prog, param_show=True))
print(convert_qprog_to_originir(prog))

machine = CPUQVM()
machine.run(prog, 1000)
results = machine.result().get_prob_list()

for key in range(2**n):
    prob = results[key]
    key_hmw = bin(key).count('1')
    if key_hmw == k:
        print(bin(key)[2::].zfill(n), prob)
```
