```python
import sympy as sp
from pyqpanda_alg.QAOA.qaoa import QAOA, p_1
from pyqpanda3.hamiltonian import PauliOperator
```

```python
vars = sp.symbols('x0:3')
f = 2*vars[0]*vars[1] + 3*vars[2] - 1
print(f)
```

```python
qaoa_f = QAOA(f)
solution_1 = [1, 0, 0]
print(qaoa_f.calculate_energy(solution_1))
```

```python
solution_2 = [0, 1, 1]
print(qaoa_f.calculate_energy(solution_2))
```

```python
ham_f = 2 * p_1(0) * p_1(1) + 3 * p_1(2) - PauliOperator({"I2" : 1.0})
qaoa_ham = QAOA(ham_f)
print(qaoa_ham.calculate_energy(solution_1))
```
