```python
from pyqpanda_alg.QAOA import default_circuits
import numpy as np
from pyqpanda3.core import QProg, CPUQVM, draw_qprog
```

```python
n = 4
k = 2
machine = CPUQVM()
prog = QProg(n)
qubits = prog.qubits()
prog << default_circuits.prepare_dicke_state(qubits, k)
print(draw_qprog(prog))
# results = machine.prob_run_list(prog, qubits)
machine.run(prog, 1000)
results = machine.result().get_prob_list(qubits)
for key in range(2**n):
    prob = results[key]
    key_hmw = bin(key).count('1')
    if key_hmw == k:
        print(bin(key)[2::].zfill(n), prob)
```
