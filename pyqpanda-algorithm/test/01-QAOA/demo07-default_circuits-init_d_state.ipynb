```python
from pyqpanda_alg.QAOA import default_circuits
from pyqpanda3.core import QProg, CPUQVM, draw_qprog
```

```python
n = 7
k = 2
machine = CPUQVM()
prog = QProg(n)
qubits = prog.qubits()
domain = [[0,1,2],[3,4,5,6]]
init_circuit = default_circuits.init_d_state(domain, k)
prog << init_circuit(qubits)
print(draw_qprog(prog))
# results = machine.prob_run_list(prog, qubits)
machine.run(prog, 1)
results = machine.result().get_prob_dict(qubits)
for key, prob in results.items():
    key = key[::-1]
    ham_1 = [key[i] for i in domain[0]].count('1')
    ham_2 = [key[i] for i in domain[1]].count('1')
    if prob > 0:
        print(key, ham_1, ham_2, prob)
```
